```python
# ====================================================================
# FINAL, GUARANTEED CODE FOR GOOGLE COLAB
# ====================================================================


# Step 1: Install all the necessary libraries
# The '!' tells Colab to run this as a command line command.
print("--- Step 1 of 4: Installing libraries... ---")
!pip install --quiet diffusers transformers accelerate


# Step 2: Import the libraries and set up the model
print("\n--- Step 2 of 4: Loading the AI model (this will be very fast!) ---")
import torch
from diffusers import AutoPipelineForText2Image
from IPython.display import display

# Because we are using Google's fast internet, we can use the best model.
pipe = AutoPipelineForText2Image.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16, # Use float16 for the powerful GPU
    variant="fp16",
    use_safetensors=True
).to("cuda") # "cuda" means we are using the FREE GPU

print("Model loaded successfully!")


# Step 3: Define your prompt
# Change this text to create anything you can imagine!
prompt = "Give me one beautiful girl where watching the  "

print(f"\n--- Step 3 of 4: Generating image for prompt: '{prompt}' ---")


# Step 4: Generate the image and display it
# This will be very fast on the GPU (10-20 seconds)
image = pipe(prompt=prompt).images[0]

print("\n--- Step 4 of 4: SUCCESS! Here is your image: ---")
display(image)
```
